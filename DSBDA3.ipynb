```python
# Import libraries
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(r"C:\Users\Ritesh Bhosale\Downloads\Sem-6\practical\DSBDA Prac\3\Iris.csv")

# Display all species
print("Unique Species:")
print(df['Species'].unique())

# Group by species
grouped = df.groupby('Species')

# Display statistics for each species
for name, group in grouped:

    print("Statistics for:", name)

    # Mean
    print("\nMean:")
    print(group.mean(numeric_only=True))

    # Standard Deviation
    print("\nStandard Deviation:")
    print(group.std(numeric_only=True))

    # Percentiles
    print("\nPercentiles:")
    print(group.quantile([0.25, 0.5, 0.75], numeric_only=True))
```
